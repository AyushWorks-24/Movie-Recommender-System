```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
df1['genres']=df1['genres'].apply(convert)
```

```python
df1['keywords']=df1['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L  
```

```python
df1['cast']=df1['cast'].apply(convert3)
```

```python
df1.head()
```

```python
def fetch_director(obj):
    for i in obj:
        if isinstance(i, dict) and i.get('job') == 'Director':
            return [i.get('name')]
    return []
```

```python
 df1['crew']= df1['crew'].apply(fetch_director)
```

```python
df1['overview']=df1['overview'].apply(lambda x:x.split())
```

```python
df1.head()
```

```python
df1['genres']=df1['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df1['keywords']=df1['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df1['cast']=df1['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df1['crew']=df1['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df1.head()
```

```python
df1['tags'] = df1['overview'] + df1['genres'] + df1['keywords'] + df1['cast'] + df1['crew']
```

```python
df1.head()
```

```python
df1.head()
```

```python
df1.drop(['crew' ], axis=1, inplace=True)
```

```python
df1.head()
```

```python
new_df=df1[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:8]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
        
```

```python
recommend('Batman Begins')
```
